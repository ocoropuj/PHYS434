## Libraries

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data

```python
mean=3
std=0.01
size = 100000

data = stats.norm.rvs(loc=mean, scale=std, size=size)
```

## Plot

```python
#Create superposed normal
x = np.linspace(2.95, 3.05, 1000)  # Create an array of x values within your desired range
pdf = stats.norm.pdf(x, loc=mean, scale=std)  # Calculate the Gaussian PDF

# Normalize the PDF to 100K
pdf = pdf * len(data) / 100000
```

```python
sns.histplot(data, bins=50, kde=False, stat='density', label=f'µ≈{data.mean():.06f}, σ≈{data.std():.06f}')
plt.plot(x, pdf, label='Gaussian PDF', linestyle='--', linewidth=2, c='orange')

plt.xlim(2.95,3.05)
plt.legend()
plt.show()
```

## Plot the log scale for y axis

```python
sns.histplot(data, bins=50, kde=True, stat='density', label=f'µ≈{data.mean():.06f}, σ≈{data.std():.06f}')
plt.plot(x, pdf, label='Gaussian PDF', linestyle='--', linewidth=2, c='orange')

plt.yscale('log')

plt.xlim(2.95,3.05)
plt.legend()
plt.show()
```
