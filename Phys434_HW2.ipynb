### Libraries and Math

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.stats import binom, poisson, norm
```

# 1) Transformation of Variables

Let $u=e^{-x} ⟹ x=-log(u)$.
$$
$$
Now $F_u(u)=ℙ(e^{-x}<=u)=ℙ(-log(u)<=x)$ Because $g$ is a uniform random variable: $F_u(u)=1+log(u) $.
$$
$$
Given that $g(u)= \frac{d}{du}F_u(u)=\frac{d}{du}(1+log(u))=\frac{1}{u} $ $\forall u\in[0,1]$.



# 2) Rolling two 6-sided dices

$a.$ Given that there is only one possible combination of the two die to roll snake eyes, and there are $36$ possible combinations this implies that $ℙ(Snake Eyes)=\frac{1}{36}$

---

$b.$ We can roll fives by: $1+4,2+3,3+2,4+1$ which are 4 combinations implies $ℙ(∑_{i=1}^{2} Dice_i = 5)=\frac{4}{36}=\frac{1}{9}$

---

$c.$ To calculate what is the probability of a snake eyes in $5$ tries is $5·ℙ(Snake Eyes)=\frac{5}{36}>\frac{4}{36}$. Therefore there is a higher probability of rolling Sneak Eyes in $5$ tries than of rolling a $5$ in the first try.



# 3) Compare Probability distribution functions

## a) Binomial $(N=10, p=0.02)$, Poisson $(λ=0.2)$, Normal$(μ=0.2, σ=4.5)$

```python
# Define the range of x values
x = np.linspace(0, 10000, 100000)
x_norm = np.linspace(-10, 10, 1000)

# Calculate the PDFs for the three distributions
N, p = 10, 0.02
lambda_poisson = 0.2
mu_normal, sigma_normal = 0.2, 0.45

binom_pmf = binom.pmf(x, 10, 0.2)
poisson_pmf = poisson.pmf(x, 0.2)
normal_pdf = norm.pdf(x_norm, 0.2, 0.45)

# Plot the PDFs
plt.figure(figsize=(10, 6))
plt.plot(x, binom_pmf, label=f'Binomial (N={N}, p={p})', alpha=1, linewidth=2)
plt.plot(x, poisson_pmf, label=f'Poisson (λ={lambda_poisson})', alpha=0.5, linewidth=5)
plt.plot(x_norm, normal_pdf, label=f'Normal (μ={mu_normal}, σ={sigma_normal})')

# Add labels and a legend
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('PDFs of Different Distributions')
plt.legend()
plt.xlim(-3,10)

# Show the plot
plt.grid(True)
print('i)')
plt.show()
```

```python
# Parameters for the distributions
N_binomial = N
p_binomial = p
lambda_poisson = lambda_poisson
mu_normal = mu_normal
sigma_normal = sigma_normal

# Calculate the statistics
mean_values = [binom.mean(N_binomial, p_binomial), poisson.mean(lambda_poisson), norm.mean(loc=mu_normal, scale=sigma_normal)]
sigma_values = [binom.std(N_binomial, p_binomial), poisson.std(lambda_poisson), norm.std(loc=mu_normal, scale=sigma_normal)]
skewness_values = [binom.stats(N_binomial, p_binomial, moments='s'), poisson.stats(lambda_poisson, moments='s'), norm.stats(loc=mu_normal, scale=sigma_normal, moments='s')]
kurtosis_values = [binom.stats(N_binomial, p_binomial, moments='k'), poisson.stats(lambda_poisson, moments='k'), norm.stats(loc=mu_normal, scale=sigma_normal, moments='k')]

# Create a DataFrame to display the statistics
data = {
    'Distribution': [f'Binomial (N={N_binomial}, p={p_binomial})', f'Poisson (λ={lambda_poisson})', f'Normal (μ={mu_normal}, σ={sigma_normal})'],
    'Mean': mean_values,
    'Sigma': sigma_values,
    'Skewness': skewness_values,
    'Kurtosis': kurtosis_values
}

df = pd.DataFrame(data)
print('ii)')
print(df)
```
